```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Graph Values
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(y)
```

#Plotting Data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title x-axis and y-axis label
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("Sin Values")
plt.title("Sine Wave")
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
#colour and symbol
plt.plot(x,y,"g.")
plt.show()
```

```python
#Two Graphs
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),"g--")
plt.plot(x,np.cos(x),"r--")
plt.show()
```

```python
#Bar Plot
fig=plt.figure() #creates empty plot
ax=fig.add_axes([0,0,0.8,1]) #creates axes
langs=["C","C++","Java","Python","PHP"]
students=[23,17,35,29,12]
ax.bar(langs,students)
plt.xlabel("Programming Languages")
plt.ylabel("Number of Students")
plt.title("Student Survey")
plt.show()
```

```python
#Pie Chart
fig1=plt.figure()
ax1=fig1.add_axes([0,0,1,1])
langs=["C","C++","Java","Python","PHP"]
students=[23,17,35,29,12]
ax1.pie(students,labels=langs,autopct="%1.1f%%")
plt.show()
```

```python
#Scatter PLot
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax2=fig2.add_axes([0,0,1,1])
ax2.scatter(x,y,color="g",label="Sine")
ax2.scatter(x,z,color="r",label="Cosine")
plt.show()
```

```python
#3D Scatter Plot
fig3=plt.figure()
ax=fig3.add_axes([0,0,1,1],projection="3d")
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap="Reds")
plt.show()
```
